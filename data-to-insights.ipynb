## Pie Chart

```python
import matplotlib.pyplot as plt

brands=["Oneplus","Apple","Samsung","Realme","POCO"]
sell=[23,21,49,64,34]
c=["black","silver","Orange","Yellow","Blue"]
#to highlight using explode
ex=[0,0.1,0,0,0]
plt.pie(sell,labels=brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=180,)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("customers.csv")
gp=df.groupby(["Gender"]).agg({"CustomerID":"count"})
print(gp)
plt.pie(gp["CustomerID"],labels=gp.index,autopct="%.2f")
plt.show()
print(df)
```

## stack plot

```python
days=[1,2,3,4,5,6,7]
NOP1=[23,27,29,37,56,76,87]
NOP2=[56,21,24,34,81,23,78]
NOP3=[38,61,23,54,28,35,86]
plt.stackplot(days,NOP1,NOP2,NOP3,colors=["red","pink","magenta"],labels=["week1","week2","week3"])
plt.legend()
plt.show()
```

```python
import pandas as pd
data=pd.read_csv("Diabetes.csv")
df=pd.DataFrame(data)
grouped=df.groupby("Age")[["Glucose","Diastolic_BP","Skin_Fold"]].agg("mean")
grouped2=grouped.head(5)
print(grouped2)
plt.stackplot(df["Age"].head(5),grouped2["Glucose"],grouped2["Diastolic_BP"],grouped2["Skin_Fold"])
plt.show()
```

## Working with Legend

```python
x=[23,27,29,37,56,76,87]
y1=[56,21,24,34,81,23,78]
y2=[38,61,23,54,28,35,86]
y3=[98,62,83,54,48,39,81]
plt.figure(figsize=[5,3])
plt.plot(x,y1,label="week1")
plt.plot(x,y2,label="week2")
plt.plot(x,y3,label="week3")
plt.legend(bbox_to_anchor=(0.8,1.2),labelspacing=2,ncols=2)
plt.show()
```
